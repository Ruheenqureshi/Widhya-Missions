```python
#importing the neccessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
#reading and importing the data
df = pd.read_csv("covid19.csv")
print('data imported succesfully')
df.head()
```

```python
#to get the daily case count
recovery_count = df['Cured'].groupby(df['ConfirmedIndianNational'], sort = False)
rec = print(recovery_count.sum())
death_count = df['Deaths'].groupby(df['ConfirmedIndianNational'],sort = False)
death = print(death_count.sum())
```

```python
recovery_count1 = df['Cured'].groupby(df['ConfirmedForeignNational'], sort = False)
rec1 = print(recovery_count1.sum())
death_count1 = df['Deaths'].groupby(df['ConfirmedForeignNational'], sort = False)
death1 = print(death_count1.sum())
```

```python
df1 = df.groupby('Date', sort = False)[['ConfirmedIndianNational','ConfirmedForeignNational',
                                      'Cured','Deaths']].sum()
```

```python
total_count = df1.sum(axis = 1)
total_count
```

```python
data = pd.DataFrame(total_count)
```

```python
data
```

```python
#plotting the data
data.plot(kind = 'bar') 
plt.title('Total cases')
plt.xlabel('Dates')  
plt.ylabel('total count')
plt.show()
```

```python
#to compute the rate of increase on each day
data1 = data.pct_change()
data1
```

```python
#getting the rate of increase for the specific columns
r = data1.loc['04/03/20': '21/03/20']
r
```

```python
 #the average rate of increse
r1 = r.mean()
r1
```
